```python
import pandas as pd
df = pd.read_csv('IMLS.csv',encoding='iso-8859-1',sep=',')
df.head()
```

```python
import plotly.plotly as py

df['text'] = df['Institution'] + '<br>Award ' + (df['AwardTotal']).astype(str)+' dollar'
limits = [(1000,5000),(5000,10000),(10000,100000),(100000,1000000),(1000000,10000000)]
colors = ["rgb(0,201,87)","rgb(0,116,217)","rgb(255,255,0)","rgb(255,128,0)","rgb(255,65,54)"]
cities = []
 
for i in range(0,3):
    lim = limits[i]
    df_sub=df[((df['AwardTotal'])>lim[0])&((df['AwardTotal'])<=lim[1])]
    city = dict(
        type = 'scattergeo',
        locationmode = 'USA-states',
        lon = df_sub['Longitude'],
        lat = df_sub['Latitude'],
        text = df_sub['text'],
        marker = dict(
            size = df_sub['AwardTotal']/500,
            color = colors[i],
            line = dict(width=0.5, color='rgb(40,40,40)'),
            sizemode = 'area'
        ),
         name ='{0} - {1}'.format(limits[i][0],limits[i][1])+'  US dollar' )
        
    cities.append(city)
for i in range(3,len(limits)):
    lim = limits[i]
    df_sub=df[((df['AwardTotal'])>lim[0])&((df['AwardTotal'])<=lim[1])]
    city = dict(
        type = 'scattergeo',
        locationmode = 'USA-states',
        lon = df_sub['Longitude'],
        lat = df_sub['Latitude'],
        text = df_sub['text'],
        marker = dict(
            size = df_sub['AwardTotal']/5500,
            color = colors[i],
            line = dict(width=0.5, color='rgb(40,40,40)'),
            sizemode = 'area'
        ),
         name ='{0} - {1}'.format(limits[i][0],limits[i][1])+'  US dollar' )
        
    cities.append(city)

layout = dict(
        title = 'Awards Amount and Awarded Institutions Distribution(1996-2003)<br>(Click legend to toggle traces)',
        showlegend = True,
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showland = True,
            landcolor = 'rgb(217, 217, 217)',
            subunitwidth=1,
            countrywidth=1,
            subunitcolor="rgb(255, 255, 255)",
            countrycolor="rgb(255, 255, 255)"
        ),
    )
plotly.tools.set_credentials_file(username='JinghanTang', api_key='Kzi5kFRGzdFAQEvHsIN0')
fig = dict( data=cities, layout=layout )
py.iplot( fig, validate=False, filename='price_amount_distribution' )
```
